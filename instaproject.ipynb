```python
def delete_words_by_char(text, char):
  """
  This function removes words containing a specific character from a string.

  Args:
      text: The original string.
      char: The character to use as a delimiter.

  Returns:
      A new string with words containing the character removed.
  """
  new_text = ""
  start_index = 0
  for i in range(len(text)):
    if text[i] == char:
      # Add the substring from the previous character to the current one (excluding char)
      new_text += text[start_index:i]
      start_index = i + 1
  # Add the last word if it doesn't contain the character
  new_text += text[start_index:]
  return new_text

text = "This is a string with some words. Here's another word"
char = ' '  # Delete words based on spaces

new_text = delete_words_by_char(text, char)
print(new_text)  # Output: ThisisastringwithsomeswordsHeresanotherword
```
